```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
review = 1000
negative_chance = 0.168
human_cost = 0.88
review_value = 10
mislabel_cost = 1.15
effectiveness = [0.9, 0.55, 0.62]
precision_loss = [0.1, 1.6, 3]
```

human_value = (negative_chance * review * review_value * effectiveness[0]) - (human_cost * review) - (negative_chance * review * mislabel_cost *precision_loss[0]) 
model1_value = (negative_chance * review * review_value * effectiveness[1]) - (negative_chance * review * mislabel_cost *precision_loss[1]) 
model2_value = (negative_chance * review * review_value * effectiveness[2]) - (negative_chance * review * mislabel_cost * precision_loss[2]) 

# Sensitivity Analysis with 100 x 100 search

```python
value = np.zeros((100,100))
for index_r, review in enumerate(np.arange(0, 1000, 10)):
    for index_c, mislabel_cost in enumerate(np.arange(0,2,0.02)):
        human_value = (negative_chance * review * review_value * effectiveness[0]) - (human_cost * review) - (negative_chance * review * mislabel_cost *precision_loss[0]) 
        model1_value = (negative_chance * review * review_value * effectiveness[1]) - (negative_chance * review * mislabel_cost *precision_loss[1]) 
        value[index_r, index_c] = human_value - model1_value
```

```python
sns.heatmap(value)
plt.title("Value difference between Human and Model 1")
plt.xlabel("Mislabel cost")
plt.ylabel("Number of reviews")
plt.xticks([25,75],["$0.5","$1.5"])
plt.yticks([25,75],[250, 750])
plt.show()
```

# Sensitivity Analysis with 10 x 10 search

```python
value = np.zeros((10,10))
for index_r, review in enumerate(np.arange(0, 1000, 100)):
    for index_c, mislabel_cost in enumerate(np.arange(0,2,0.2)):
        human_value = (negative_chance * review * review_value * effectiveness[0]) - (human_cost * review) - (negative_chance * review * mislabel_cost *precision_loss[0]) 
        model1_value = (negative_chance * review * review_value * effectiveness[1]) - (negative_chance * review * mislabel_cost *precision_loss[1]) 
        value[index_r, index_c] = human_value - model1_value
```

```python
sns.heatmap(value, annot = True, fmt=".0f")
plt.title("Value difference between Human and Model 1")
plt.xlabel("Mislabel cost")
plt.ylabel("Number of reviews")
plt.xticks([2,7],["$0.4","$1.6"])
plt.yticks([2,7],[200, 700])
plt.scatter(value[3][5],value[5][3], c = "green", s = 50)
plt.show()
```
